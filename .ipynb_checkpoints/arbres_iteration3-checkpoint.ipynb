```python
#En vous servant de l’annexe A, identifiez toutes les essences “non désirées” à Grenoble. Quelle est la proportion “essences désirées” vs “non désirées” ?

import numpy as np
df["Espece botanique"] = df["GENRE_BOTA"]+ ' '+ df["ESPECE"]
botanique_temps = df[["ANNEEDEPLANTATION","Espece botanique"]].dropna().value_counts()
botanique_temps_nombreplanta = botanique_temps.reset_index(name="nombre de plantations")
espece_indésirable = ["Acer saccharinum","Albizia jullibrissin","Catalpa bignonoides","Sophora japonica","Ginkgo biloba","Gleditsia triacanthos","Koelreuteria paniculata","Laburnum anagynoides","Paulownia tomentosa"]
tableau = botanique_temps_nombreplanta.pivot(index = "ANNEEDEPLANTATION", columns = "Espece botanique",values="nombre de plantations")

total_indesirable = 0
total_desirable = 0
for col in tableau:
    if col in espece_indésirable:
        total_indesirable +=tableau[col].sum()
    else:
        total_desirable+=tableau[col].sum()

total = np.nansum(tableau.values) #retire les "nan"

proportion_indesirable = (total_indesirable/total)*100
proportion_desirable = (total_desirable/total)*100
print(f"il y a {proportion_indesirable}% d'especes indesirables")
print(f"il y a {proportion_desirable}% d'especes désirables")
print(total_indesirable)
print(total_desirable)
```
