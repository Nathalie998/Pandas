```python
import csv
with open('data/trees.csv', newline='',encoding = 'utf-8-sig') as f:
    reader = csv.reader(f)
    data = list(reader)
    
```

```python
#afficher la deuxieme colonne de chacune de ces lignes:
for i in range(2):
    print(data[i][1])
```

```python
#extraire l'information concernant l'année de plantation pour l'ensemble des lignes(rows) et 
#stocker ces infos dans une autre variable de type List

annee_de_plantation = []

header = data[0]

index_annee = header.index("ANNEEDEPLANTATION")

for row in data[1:]:
    annee_de_plantation.append(row[index_annee])
print(annee_de_plantation)

```

```python
#combien d'arbres sont recensés dans ce jeu de données?

l = len(annee_de_plantation)
print("il y a ", l, "arbres")
```

```python
#Pour combien d’arbres manque-t-il l’information concernant la date de plantation ? 
#(Vous pouvez essayer de le faire d’au moins deux manières différentes.)

arbres_sans_info_date=annee_de_plantation.count('')
print("il y a",arbres_sans_info_date,"arbres sans information concernant la date de plantation")
```

```python
#Combien d’arbres ont été plantés l’année de votre naissance ? 
liste_arbres_depuis_naissance = []
for annee in annee_de_plantation:
    if annee>='1998':
        liste_arbres_depuis_naissance.append(annee)
#print(liste_arbres_depuis_naissance)
nombre_arbre = len(liste_arbres_depuis_naissance)
print("il y a",nombre_arbre,"plantés depuis ma naissance")
```

```python
#Quelle est la plus ancienne année de plantation recensée dans ce dataset ? La plus récente ? 

annee_sans_header = annee_de_plantation[1:]
annee_plus_ancienne = min(int(a) for a in annee_sans_header if a.strip() != '')

print("la valeur la plus ancienne est", annee_plus_ancienne)

annee_plus_recente = max(int(a) for a in annee_sans_header if a.strip() != '')

print("la valeur la plus récente est", annee_plus_recente)
```

```python
#Combien d'arbres ont été plantés par année
liste_annees=[]
for i in range(1950,2018):
    liste_annees.append(i)

#initialisation d'un dictionnaire
compte_arbres = {annee : 0 for annee in liste_annees}
    
for annee in annee_de_plantation[1:]:
    if annee.strip() != '':
        annee = int(annee)
        if annee in compte_arbres:
            compte_arbres[annee] += 1 
            
print(compte_arbres)

#for annee in liste_annees:       
#    print(f"pour l'annee {annee} il y a {compte_arbres[annee]} arbres")  
```

```python
#Combien d’arbres ont été plantés en moyenne chaque année ?   Moyenne d’une liste de deux manières : 

nombre_darbres = 0
nombre_dannees = 0

for nombre in compte_arbres.values():
    nombre_darbres += nombre

for annee in compte_arbres.keys():
    nombre_dannees += 1
#print(len(set(compte_arbres.keys())))

moyenne = nombre_darbres/nombre_dannees
print(f"En moyenne, il y a {moyenne} d'arbres plantés chaque années")
```

```python
#Stocker conjointement l’année de plantation et le nombre d’arbres plantés dans un tuple.
#Les tuples seront stockés dans une liste (ex : [ (‘1987’, 771), (‘1988’, 266), ….. ] )

liste_dannees_et_darbres = list(compte_arbres.items())
print(liste_dannees_et_darbres)
```
